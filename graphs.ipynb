```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

%matplotlib inline
```

## Spline Points

```python
csv_dir = "./dev-files/splines/"

def plotPoints(data_id: int, path = ""):
	if not path:
		csv_path_1 = csv_dir + f"spline{data_id}.csv"
	else:
		csv_path_1 = path

	data1 = pd.read_csv(csv_path_1)

	fig, (ax1) = plt.subplots(1)
	size = 3

	def plotPoints(ax: matplotlib.axes.Axes, x, y, title, z: pd.Series = pd.Series()):
		ax.scatter(x, y, s=size)
		if z.empty:
			for i, d in enumerate(z):
				ax.annotate(d, [x[i], y[i]], size=5, color="gray")
		# ax.plot(x, y, ms=size)
		ax.set_xlim(0, 6)
		ax.set_ylim(0, 6)
		ax.set_title(title)

	if "s" in data1.columns:
		plotPoints(ax1, data1["x"], data1["y"], f"spline{data_id}", data1["s"])
	else:
		plotPoints(ax1, data1["x"], data1["y"], f"spline{data_id}")

	plt.show()


# plotPoints(0, "./test.csv")
# plotPoints(0, "./test1.csv")
plotPoints(0)
plotPoints(1)
plotPoints(2)
```

## Trajectory Path

```python
csv_dir = "./dev-files/paths/"

def plotPathCSVs(path_id: int):
	csv_path_1 = csv_dir + f"path{path_id}-dis.csv"
	csv_path_2 = csv_dir + f"path{path_id}-vel.csv"
	csv_path_3 = csv_dir + f"path{path_id}-ang_vel.csv"
	csv_path_4 = csv_dir + f"path{path_id}-accel.csv"
	csv_path_5 = csv_dir + f"path{path_id}-k.csv"

	data1 = pd.read_csv(csv_path_1, index_col=0)
	data2 = pd.read_csv(csv_path_2, index_col=0)
	data3 = pd.read_csv(csv_path_3, index_col=0)
	data4 = pd.read_csv(csv_path_4, index_col=0)
	data5 = pd.read_csv(csv_path_5, index_col=0)

	fig, ((ax1, ax2), (ax3, ax4), (ax6, ax5)) = plt.subplots(3, 2)
	size = 3

	def plotData(ax: matplotlib.axes.Axes, data, title):
		for col in data.columns:
			ax.scatter(data.index, data[col], s=size, label=col)
		ax.set_title(title)
		ax.legend()

	plotData(ax1, data1, "distance")
	plotData(ax2, data2, "velocity")
	plotData(ax3, data3, "angular vel")
	plotData(ax4, data4, "acceleration")
	plotData(ax5, data5, "curvature")
	print(data1.index[-1])

	plt.show()


plotPathCSVs(0)
plotPathCSVs(1)
plotPathCSVs(2)
plotPathCSVs(3)
plotPathCSVs(4)

# Old
# 3.355
# 8.88
# 4.885
# 8.995
# 14.84

# Slightly faster calculation
# 3.43
# 8.99
# 4.92
# 9.015
# 14.94
```

## Polygon

```python
csv_dir = "./dev-files/polygons/"

def plotPoints(data_id: int, path = ""):
	csv_path_poly = csv_dir + f"polygon{data_id}-poly.csv"
	csv_path_points = csv_dir + f"polygon{data_id}-points.csv"

	data_poly = pd.read_csv(csv_path_poly)
	data_points = pd.read_csv(csv_path_points)

	fig, (ax1) = plt.subplots(1)

	ax1.fill(data_poly["poly_x"], data_poly["poly_y"])
	ax1.scatter(data_points["point_x"], data_points["point_y"], s=1, c=data_points["is_in"])

	plt.show()


plotPoints(0)
plotPoints(1)
plotPoints(2)
```
